Data Preprocessing Procedure:
To prepare the dataset for training, I used PyTorch’s transforms module to apply a sequence of preprocessing steps for both the training and validation/test sets.

Image Resizing:
For resizing, I first used transforms.Resize(256) to make the shortest side of each image 256 pixels while keeping the aspect ratio. Then, I used transforms.CenterCrop(224) to crop the image to 224×224 pixels centered around the image. This standard size (224×224) matches what most pretrained CNNs like ResNet18 expect as input.

Input Tensor Size:
The final input tensor size is (3, 224, 224), where 3 is the number of color channels (RGB). I chose this size because it's the default input size for most pretrained models available in PyTorch's torchvision library. It ensures compatibility and saves time tuning dimensions manually.

Data Augmentation:
Yes, I applied data augmentation to the training set to improve generalization and reduce overfitting. I used:
transforms.RandomResizedCrop(224) – crops a random portion of the image and resizes it to 224×224.
transforms.RandomHorizontalFlip() – flips the image horizontally with a 50% chance.
I did not apply these augmentations to the validation and test sets. For those, I only used resizing and center cropping to ensure consistent and fair evaluation.


Summary:
Data augmentation helped my model learn better by providing more diverse training examples, and resizing to 224×224 allowed me to use a pretrained model efficiently. These preprocessing steps helped improve my model’s accuracy while keeping the training stable.
